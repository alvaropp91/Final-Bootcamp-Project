# Model

APPLE

```python
apple_data.Date = apple_data.Date.astype("datetime64")
apple_data["year"] = pd.DatetimeIndex(apple_data.Date).year.astype("object")
apple_data["month"] = pd.DatetimeIndex(apple_data.Date).month.astype("object")
apple_data["day"] = pd.DatetimeIndex(apple_data.Date).day.astype("object")
apple_data.Date = apple_data.Date.astype("object")
apple_data = apple_data.drop(columns = ["Date"])
```

```python
apple_data
```

```python
apple_data.dtypes
```

```python
X_apple = apple_data.drop(['High'], axis=1)
y_apple = apple_data['High']
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X_apple_train, X_apple_test, y_apple_train, y_apple_test = train_test_split(X_apple, y_apple, test_size=0.20, random_state=42)
```

```python
num_train_apple = X_apple_train.select_dtypes(np.number)
num_test_apple = X_apple_test.select_dtypes(np.number)
cat_train_apple = X_apple_train.select_dtypes(object)
cat_test_apple = X_apple_test.select_dtypes(object)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_apple = MinMaxScaler().fit(num_train_apple)
num_train_scaled_apple = scaler_apple.transform(num_train_apple)
num_test_scaled_apple = scaler_apple.transform(num_test_apple)
cols_apple = scaler_apple.get_feature_names_out(input_features = num_train_apple.columns)
num_train_scaled_apple = pd.DataFrame(num_train_scaled_apple, columns=cols_apple)
num_test_scaled_apple = pd.DataFrame(num_test_scaled_apple, columns=cols_apple)
```

```python
encoder_apple = OneHotEncoder(handle_unknown ='error', drop='first').fit(cat_train_apple)
cat_train_encoded_apple  = encoder_apple.transform(cat_train_apple).toarray()
cat_test_encoded_apple = encoder_apple.transform(cat_test_apple).toarray()
cols_apple = encoder_apple.get_feature_names_out(input_features = cat_train_apple.columns)
cat_train_encoded_apple = pd.DataFrame(cat_train_encoded_apple, columns = cols_apple)
cat_test_encoded_apple = pd.DataFrame(cat_test_encoded_apple, columns = cols_apple)
```

```python
X_apple_train = pd.concat([num_train_scaled_apple, cat_train_encoded_apple], axis = 1)
X_apple_test = pd.concat([num_test_scaled_apple, cat_test_encoded_apple], axis = 1)
```

```python
import math
from sklearn import linear_model
LR_apple = linear_model.LinearRegression()
LR_apple.fit(X_apple_train,y_apple_train)
```

```python
from sklearn.metrics import r2_score
predictions_test_apple = LR_apple.predict(X_apple_test)
r2_score(y_apple_test, predictions_test_apple)
```

```python
predictions_train_apple = LR_apple.predict(X_apple_train)
r2_score(y_apple_train, predictions_train_apple)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
mse_train_apple=mean_squared_error(y_apple_train,predictions_train_apple)
print('The mean squared error train is ', mse_train_apple)
rmse_train_apple = np.sqrt(mean_squared_error(y_apple_train,predictions_train_apple))
print ('The root-mean-square error train is', rmse_train_apple)
mae_train_apple = mean_absolute_error(y_apple_train, predictions_train_apple)
print ('The mean absolute error train is', mae_train_apple)
mse_test_apple =mean_squared_error(y_apple_test,predictions_test_apple)
print('The mean squared error test is ', mse_test_apple)
rmse_test_apple = np.sqrt(mean_squared_error(y_apple_test,predictions_test_apple))
print ('The root-mean-square error test is', rmse_test_apple)
mae_test_apple = mean_absolute_error(y_apple_test, predictions_test_apple)
print ('The mean absolute error test is', mae_test_apple)
```

```python
X_apple_num = X_apple.select_dtypes(np.number)
X_apple_cat = X_apple.select_dtypes(object)
```

```python
X_cat_encoded_apple  = encoder_apple.transform(X_apple_cat).toarray()
cols_apple1 = encoder_apple.get_feature_names_out(input_features = X_apple_cat.columns)
X_cat_encoded_apple = pd.DataFrame(X_cat_encoded_apple, columns = cols_apple1)
```

```python
X_apple_scaled_num = scaler_apple.transform(X_apple_num)
cols_apple2 = scaler_apple.get_feature_names_out(input_features = X_apple_num.columns)
X_scaled_apple_num = pd.DataFrame(X_apple_scaled_num, columns=cols_apple2)
```

```python
X_apple = pd.concat([X_scaled_apple_num, X_cat_encoded_apple], axis = 1)
```

```python
predictions_apple = LR_apple.predict(X_apple)
```

```python
X_apple["Predictions"] = predictions_apple
```

WEB SCRAPING EXTRACTING THE COMPONENTS OF NASDAQ (FROM WIKIPEDIA)

```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Nasdaq-100"

response = requests.get(url)
html_content = response.text

soup = BeautifulSoup(html_content, "html.parser")

table = soup.find("table", attrs={"id": "constituents"})
```

```python
table_data = []
for row in table.find_all("tr"):
    row_data = []
    for cell in row.find_all(["th", "td"]):
        row_data.append(cell.text.strip())
    if row_data:
        table_data.append(row_data)
```

```python
df = pd.DataFrame(table_data[1:], columns=table_data[0])
print(df)
```
